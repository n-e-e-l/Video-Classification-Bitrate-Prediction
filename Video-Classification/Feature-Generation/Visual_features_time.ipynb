```
from scenedetect.detectors import ContentDetector
from scenedetect import VideoManager
from scenedetect import SceneManager
import os
import cv2
import numpy as np
i = 0
directory = r'drive/My Drive/five'
for filename in os.listdir(directory):
    i = i + 1
    if filename.endswith(".mp4"):
        temp_name=os.path.join(directory, filename)
        #video_path = r'H:\Thesis_data\test\33_nonanimatedvalidation.mp4'
        video_manager = VideoManager([temp_name])
        scene_manager = SceneManager()
        scene_manager.add_detector(ContentDetector(threshold=30))
        video_manager.start()
        scene_manager.detect_scenes(frame_source=video_manager)
        scene_list = scene_manager.get_scene_list(video_manager.get_base_timecode())
        time = []
        for scene in scene_list:
            start, end = scene
            #print(f'{start.get_seconds()} - {end.get_seconds()}')
            temp = float(end.get_seconds())-float(start.get_seconds())
            time.append(temp)
            #print(float(end.get_seconds())-float(start.get_seconds()))
        #print(sum(time)/len(time))
        temp_time=sum(time)/len(time)
        df.loc[len(df.index)] = [filename, temp_time]
    else:
        continue
```

```
df = df.iloc[3:]
```
